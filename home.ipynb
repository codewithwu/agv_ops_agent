```python
# ============================================
# AGV 基本操作手册 - RAG 完整流程演示
# 使用已测试通过的 VectorStoreService
# ============================================

import sys
sys.path.insert(0, "/home/cooper/githubProjects/agv_ops_agent")

from src.services.vectorstore import VectorStoreService

# -------- 1. 初始化 VectorStoreService --------
vs_service = VectorStoreService(
    embedding_provider="ollama",
    collection_name="agv_manual",
    chunk_size=300,
    chunk_overlap=50,
)
print("VectorStoreService 初始化完成")

# -------- 2. 加载并存储文档 --------
doc_ids = vs_service.add_documents(
    file_path="/home/cooper/githubProjects/agv_ops_agent/01_AGV基本操作手册.md",
    metadata={"source": "AGV基本操作手册", "version": "1.0"}
)
print(f"文档已加载并存储，块数量: {len(doc_ids)}")

# -------- 3. 检索测试 --------
vs = vs_service.get_vectorstore()

# 测试1: AGV无法启动怎么办
print("\n" + "=" * 50)
print("测试1: AGV无法启动怎么办")
print("=" * 50)
results = vs.similarity_search(query="AGV无法启动怎么办", k=3)
for i, doc in enumerate(results):
    print(f"\n--- 结果 {i+1} ---")
    print(doc.page_content[:300] + "..." if len(doc.page_content) > 300 else doc.page_content)
    print(f"元数据: {doc.metadata}")

# 测试2: 手动模式如何使用
print("\n" + "=" * 50)
print("测试2: 手动模式如何使用")
print("=" * 50)
results = vs.similarity_search(query="手动模式如何使用", k=3)
for i, doc in enumerate(results):
    print(f"\n--- 结果 {i+1} ---")
    print(doc.page_content[:300] + "..." if len(doc.page_content) > 300 else doc.page_content)
    print(f"元数据: {doc.metadata}")

# 测试3: 带分数检索 - 电池维护
print("\n" + "=" * 50)
print("测试3: 带分数检索 - 电池电量低于20%")
print("=" * 50)
results_with_scores = vs.similarity_search_with_score(query="电池电量低于20%怎么办", k=3)
for doc, score in results_with_scores:
    print(f"\n[分数: {score:.4f}]")
    print(doc.page_content[:300] + "..." if len(doc.page_content) > 300 else doc.page_content)
```
